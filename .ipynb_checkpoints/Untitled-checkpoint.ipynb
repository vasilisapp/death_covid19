```python
def difference(category, df):
    df['difference_' + category] = df[category].shift(-1) - df[category]
    return df 
X = difference('total', X)
X = difference('by_age', X)
X = difference('disability', X)
X = difference('loss_of_breadwinner', X)
X = difference('for_years_of_seniority', X)
X = difference('social', X)
X = difference('retired_judges', X)
X = difference('military', X)
X = X[:-1]
print(X[['difference_total', 'difference_by_age', 'difference_disability']])
```

```python
X.difference_total[0:2].sum()
```

```python
years_index = ['2015', '2016', '2017', '2018', '2019', '2020', '2021']


#sum_year_difference_total = pd.DataFrame().reindex_like(risky_r)
sum_year_difference_total = [0,0,0,0,0,0,0]
sum_year_difference_total[0] = X.difference_total[0:2].sum()
sum_year_difference_total[1] = X.difference_total[2]
sum_year_difference_total[2] = X.difference_total[3]
sum_year_difference_total[3] = X.difference_total[4:8].sum()
sum_year_difference_total[4] = X.difference_total[8:12].sum()
sum_year_difference_total[5] = X.difference_total[12:16].sum()
sum_year_difference_total[6] = X.difference_total[16].sum()

sum_year_difference_age = [0,0,0,0,0,0,0]
sum_year_difference_age[0] = X.difference_by_age[0:2].sum()
sum_year_difference_age[1] = X.difference_by_age[2]
sum_year_difference_age[2] = X.difference_by_age[3] 
sum_year_difference_age[3] = X.difference_by_age[4:8].sum()
sum_year_difference_age[4] = X.difference_by_age[8:12].sum() 
sum_year_difference_age[5] = X.difference_by_age[12:16].sum()
sum_year_difference_age[6] = X.difference_by_age[16]

df = pd.DataFrame({'sum_year_difference_total': sum_year_difference_total, 
                   'sum_year_difference_age': sum_year_difference_age,
                   'years_index' : years_index})

df = df.set_index(pd.to_datetime(years_index, format = "%Y").to_period('Y'))
df = df.drop(['years_index'], axis=1)
print(df)
```

```python
import matplotlib.pyplot as plt

df.plot.bar(legend=True, alpha=0.5, color=['teal','goldenrod']);
plt.grid(color='#95a5a6', linestyle='--', linewidth=1, axis='y', alpha=0.5)
xc = 2.5
plt.axvline(x=xc, color='k', linestyle='--');
#plt.text(x = 2 + 0.1  , y = 150, s = 'пенсiйна реформа 2017', size = 10)
plt.legend(labels=['пенсiйна реформа 2017', 'усього,тис.', 'за віком,тис.']);
plt.xticks(range(0, 7), df.index, rotation=0);
```

```python
def difference_year(category, df):
    #df['12_month_' + category] = df[category].cumsum().shift(+4).ffill()
    df['year_' + category] = df[category].rolling(min_periods=4, window=4).sum()
    return df 
for category in cols:
    df_2018_2021 = difference_year(category, df_2018_2021)
    
print(df_2018_2021[['year_difference_total', 'year_difference_by_age', 'year_difference_disability']])
```
